# Sweetness Prediction using both dataset 

```r
# loading libraries 
library(glmnet)
library(caTools)
set.seed(100)
```

```r
# Reading data
sp <- read.csv("C:/Users/Shubham Shukla/Desktop/internship/new/sweetness_prediction/ecfp.csv")
sp1 <- read.csv("C:/Users/Shubham Shukla/Desktop/internship/new/sweetness_prediction/new_data/ecfp_new.csv")
```

```r
#converting sweetness value of sp in log 
sp$Sweetness <- log10(sp$Sweetness)
names(sp)[2] <- "logSw"
```

```r
#Combining both dataset
sp2 <- rbind(sp,sp1)
dim(sp2)
```

```r
# Splitting sp2 data into training and testing set

split <- sample.split(sp2$logSw, SplitRatio = 0.75)
train <- subset(sp2, split == T, select = -c(smiles))
test <- subset(sp2, split == F, select = -c(smiles))
```

```r
dim(test)
```

# Using MLR 

```r
model1 <- lm(logSw ~., data = train)
summary(model1)
```

```r
pred1 <- predict(model1, newdata = test)
```

```r
length(pred1)
```

```r
sse <- sum((test$logSw - pred1)^2)
sst <- sum((test$logSw - mean(test$logSw))^2)

(sse-sst)/sse
```
